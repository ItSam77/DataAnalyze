```python
import pandas as pd
```

```python
data =  pd.read_csv('social_media_dataset.csv') ##Import csv files to colab
data.head(17)
```

```python
data.groupby('Platform').sum().head()
```

```python
total_data = data.groupby('Platform').mean() ##Group by platform
total_data.head() ##Menampilkan tabel setelah di rata2kan
```

```python
criteria = ['User Engagement','Reach and Popularity', 'Trustworthiness', 'Ease of Sharing']
bobot = [0.4,0.3,0.2,0.1]
```

```python
def norm(total_data): ##Normalisasi
  for i in criteria:
    total_data[i] = total_data[i]/total_data[i].max() ##Nilai data/ nilai maksimal pada kriteria

  return total_data
```

```python
norm_data = norm(total_data)
norm_data.head() ##Menampilkan tabel setelah di normalisasikan
```

```python
def saw(total_data,bobot):
  total_data = norm(total_data)
  hasil = total_data.mul(bobot,axis=1).sum(axis=1)
  return hasil
```

```python
hasil_saw = saw(total_data,bobot)
print("Hasil SAW =")
for platform, nilai in hasil_saw.items():
    print(f"{platform}: {nilai:.3f}")

```

```python
hasil_saw_sort = sorted(hasil_saw.items(), key=lambda x: x[1],reverse = True)  # Sort by value
print("Rank setiap social media dengan diurutkan hasil SAW =")
for i,item in enumerate(hasil_saw_sort):
    print(f"Social media ke-{i+1} = {item[0]}")
```
